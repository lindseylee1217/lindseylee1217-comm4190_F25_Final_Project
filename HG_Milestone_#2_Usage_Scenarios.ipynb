# Usage Scenarios

## Usage Scenario #1 (creators)

### Short description  
Extract repeated images, metaphors and tones from a week's creative fragments, helping **writers** organize their writing direction.

### Narrative of scenario  
Each week, the writer opens **Looking Glass** and imports their creative fragments—scenes, paragraphs, and imagery notes. The LLM analyzes the text to identify recurring imagery, tones swifts. Then reflects how these patterns shape the writer’s thematic concerns over time. It constructs an imagery map, which like a graph that node represents a recurring motif or emotional tone, and edges show where these motifs frequently appear together in the writer’s work. The system refers two short quotes from the writer’s own text as evidence, and offers three guiding questions to help clarify next week’s writing direction. Finally, the writer chooses the thread that feels most attracting, names it, and saves it as a next-week writing focus card for ongoing development.

### Step by step breakdown  
```{mermaid}
sequenceDiagram
    actor U as Writer
    participant R as Looking Glass
    participant L as LLM
    U->>R: Import fragments and imagery notes
    R->>L: Integrate recurring images and tone shifts
    L-->>R: Generate  map + quotes + guiding questions
    R-->>U: Present options for next-week focus
    U->>R: Pick one focus and name it
    R-->>U: Save and export as Markdown
```

### Data Description

#### model training Data

- Labeled journal entries/Emotion classification dataset
- HappyDB (GitHub)
- Public forum posts (Reddit)

#### Needed from user 
- imagery cards
- vignettes
- writing sketches
- Or anything else they wrote in the week 

#### Given to LLM
user's writing content + imagery tags  

#### Created
- Structured imagery map  
- Three guiding questions  
- Next-week writing thread card  

### Evaluation

**Success Criteria**  
- Imagery extraction matches original tone  
- User agrees with the thread card  

**Possible Issue**  
- Fragments vary greatly in style  
- The narrative emotions in a novel excerpt are taken as the author's own emotions
- The metaphor of being treated literally




## Usage Scenario #2 (professionals)

### Short description  
Detecting anxiety patterns across work notes and journals of **working professionals**, reflecting stress and providing actionable self-management prompts.

### Narrative of scenario  
A new analyst at quarter-end uploads their meeting notes, task lists, email drafts and a few personal journal entries into **Looking Glass**. The system analyzes the entries and detects increased anxiety terms from Wednesday to Friday and correlates them with peak email activity. It builds a stress pattern map (graph) showing when anxiety peaks and what triggers it. Then it quotes two short excerpts as evidence, then generates two reflection questions and suggests some concrete boundary-setting actions the user can apply in the coming week.

### Step by step breakdown  
```{mermaid}
sequenceDiagram
    actor U as Analyst
    participant R as Looking Glass
    participant L as LLM
    U->>R: Import notes and diaries
    R->>L: Detect deadline-related stress patterns
    L-->>R: Triggers + summary + questions + quotes
    R-->>U: Show triggers and ask for a small change
    U->>R: Confirm change and rationale
    R-->>U: Save plan for next review
```

### Data Description

#### model training Data

- Labeled journal entries/Emotion classification dataset
- HappyDB (GitHub)
- Public forum posts (Reddit)

#### Needed from user 
- meeting notes  
- task lists  
- personal journal entries  
- Or anything else they wrote in the week 

#### Given to LLM
user's writing content + text segments labeled with timestamp  

#### Created
- Graph of stress triggers  
- Two reflection questions  
- One small actionable change plan  

### Evaluation

**Success Criteria**  
- Triggers align with the actual timeline  
- User feels the suggested change is concrete and doable  

**Possible Issue**  
- The week consists almost entirely of work emails, with almost no personal expression. Emotional visibility is low
- Arcasm, jokes, and habitual phrases are misread as positive emotions. e.g., "Great, another five-hour meeting today"
- Meeting notes are written in third-person, thus emotional signals even harder to detect
