**Idea Pitch: Looking Glass**
- Have you ever felt that life moves so fast that your thoughts never have the chance to be heard? Every week, we write so many words in journals, messages, and notes, yet we rarely look back to see what they reveal about us. Looking Glass helps you do just that. It uses LLMs to analyze writing over time and notices subtle patterns. It finds the emotions that rise and fade, the topics that return, the words that you repeat without realizing it. Then, it turns all of this into reflections. You can read a clear summary of your emotional and thematic shifts, and talk with the system as it asks gentle questions to help you think deeper. It is a quiet space that slows the rhythm of your thoughts and lets you meet them again. In a world that moves too fast, it gives you the time to pause, to listen, and to remember who you are becoming.

**Target Users**
- Students/professionals
- Writers (personal or professional)
- Individuals seeking to improve their mental health
- Individuals who journal

**Detailed Summary of Application Use**
- Users open the app and create a new space/category to import journals and notes, or they can directly write or paste in the box.
- User is prompted to select a preference of LLM reflective style and tone—which also adapts over time based on user data entry.
- LLM analyzes textual data entry to identify tone, emotions, and how it may reflect on the context and the individual. It generates an individualized reflection of a specific textual data entry and themes of the user’s writing over a period of time. It also performs in-line summary and commentary of tones, emotions, unraveling themes in writing directly on the journal entry.
- Users may then review and use this generated output and summary to:
    - Reflect and understand the user’s emotions and tone
    - Create and keep track of goals for their personal growth
    - Read LLM-generated emotion trajectories across journal entries over time
    - View an easy emotion graph that shows how their feelings change and repeat over time
- The LLM then prompts the user if they would like to engage in talk therapy—where users may opt-in or opt-out—to further explore their emotions with LLM-guided questions. 
- Users' data stay locally and can be cleared with one tap.

**Data needed**
- Training data: journal entries, emails, and other forms of writing from a variety of sources
- Public journal entries database with labeled emotions: https://www.kaggle.com/datasets/madhavmalhotra/journal-entries-with-labelled-emotions
- Emotions dataset: https://www.kaggle.com/datasets/praveengovi/emotions-dataset-for-nlp/ GitHub - megagonlabs/HappyDB: A corpus of 100,000 happy moments
- Scrape data from public site like reddit (https://www.reddit.com/r/offmychest/, r/selfimprovement)

**Role of LLM**
- Identify tones, emotions and recurring topics in each individual journal entry and tracks how entries shift over time, summarizing reflections in a paragraph summary and a visual graph that trajects user’s emotions (e.g. whether the user’s expression of positive emotions has increased).
- Generate questions that invite user to explore the context, reason, and implications of their feelings and actions expressed in their journal entry.
- Adapts its tone of feedback and reflection based on user input and preferences.

**How does our product differ from existing non-LLM powered systems in the domain and from just using a general purpose LLM for the task**
- Our product is distinguished from non-LLM powered systems because:
    - LLMs can identify emotions from beyond hard-coded key words and inputs—LLMs can infer emotion given their rich training data
    - LLMs can respond personally and individually to a user given the context of their interaction
    - LLMs adapt the prompts and examples to user language style and preferred depth and tones
- Our product is distinguished from using general purpose LLMs because:
    - Your data and writing will have memory within the platform, specific to each user. Data will not be shared across users. Thus, there is a longitudinal analysis of user data.
    - Tailored for in-line textual analysis.
    - LLM carries a different tone from generic LLMs—adapted from user input and preferences gauged at the beginning.
